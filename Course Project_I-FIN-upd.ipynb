```python
countries = df[0:10]['Country'].sort_index(ascending=False)
HS = df[0:10]['Happiness Score'].sort_values(ascending=True)
y_coord = np.arange(len(countries))
```

```python
plt.barh(y=y_coord, width = HS, height=0.8, label = 'Happiness Score')
plt.yticks(y_coord, countries)
plt.xticks(np.arange(13))
plt.xlabel('Hapiness Score')
plt.ylabel('Country')
plt.title('Top 10 countries by Happiness Score')
plt.legend(loc='center right')
plt.show()
```

```python
#C-2. Modify the bar chart obtaines to turn into a stacked bar chart where the overall happiness score is divided into the seven parts corresponding to the columns

Eco = df[0:10]['Economy']
Fam = df[0:10]['Family']
Hea = df[0:10]['Health']
Fre = df[0:10]['Freedom']
Gen = df[0:10]['Generosity']
Cor = df[0:10]['Corruption']
Dys = df[0:10]['Dystopia']

plt.barh(y_coord, Eco, color='red', edgecolor="black", label='Economy')
plt.barh(y_coord, Fam, color='blue', left=Eco, edgecolor="black", label='Family')
plt.barh(y_coord, Hea, color='yellow', left=Eco+Fam, edgecolor="black", label='Health')
plt.barh(y_coord, Fre, color='green', left=Eco+Fam+Hea, edgecolor="black", label='Freedom')
plt.barh(y_coord, Gen, color='orange', left=Eco+Fam+Hea+Fre, edgecolor="black", label='Generosity')
plt.barh(y_coord, Cor, color='black', left=Eco+Fam+Hea+Fre+Gen, edgecolor="black", label='Corruption')
plt.barh(y_coord, Dys, color='violet', left=Eco+Fam+Hea+Fre+Gen+Cor, edgecolor="black", label='Dystopia')

plt.yticks(y_coord, countries)
plt.xticks(np.arange(12))
plt.xlabel('Hapiness Score')
plt.ylabel('Country')
plt.title('Top 10 Countries by Happiness Score in details')
plt.legend(loc='center right')
plt.show()
```

```python
#C-3. Obtain the same stacked horizontal bar chart for all countries from the region Africa.

Africa=df[df['Region'] == 'Africa'].sort_values('Happiness Score', ascending=True)
countries_Africa = Africa['Country']
y_coord_Africa = np.arange(len(countries_Africa))
```

```python
#G. Probabilities. Compute the probability that randomly selected country with Happiness score over 6.0 is from Western Europe
happy_countries = df[df['Happiness Score']>6]
```

```python
import random 
region = happy_countries['Region']
experiments = 10000
counter = 0

for i in range(experiments):
    country = region[random.randrange(len(region))]
    if country == 'Western Europe':
         counter += 1

counter/experiments

#Answer: the probability is about 0.38
```
